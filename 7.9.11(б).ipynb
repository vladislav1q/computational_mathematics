```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x_axis=[]
y_axis=[]

for i in np.arange(-2, -5, -0.1):
    h = 0.25*10**(i)
    result = 0
    for x in np.arange(1, 10, h):
        result += h*(np.exp(-x*x)*np.sin(x))
    x_axis.append(h)
    y_axis.append(result)

plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(x_axis, y_axis, c='red')
plt.plot([0.0,0.0025], [0.129738, 0.129738], c='green')
plt.xlabel('h - step to calculate the integral')
plt.ylabel('result for calculating the integral')
```

### Вывод

Видно, что при уменьшении шага интегрирование методом прямоугольников дает линейное увеличение точности результата. Заданная точность 10^-4 достигается при шаге 0,25*10^-5, что было показано в расчетах.
