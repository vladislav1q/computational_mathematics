```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Прежде было показано, что исходная задача:
$$y"=ysinx$$

$$0\leq x \leq 1$$
$$y(0)=0, y'(0)=1$$

Сводится заменой к следущей системе:

$$v_{n+1}=v_n+hy_nsinx_n$$
$$y_{n+1}=y_n+hv_n$$

$$0\leq x \leq 1$$
$$x_0=y_0=0, v_0=1$$

Используя начальные данные построим приближенное решение задачи коши при заданном шаге h.

```python
h = 0.0001

x = []
v = []
y = []

x.append(0)
y.append(0)
v.append(1)

for i in np.arange(0.0,1.0+h,h):
    x.append(i)
    y.append(y[-1]+h*v[-1])
    v.append(v[-1]+h*y[-1]*np.sin(x[-1]))

a, axes = plt.subplots(1,2, figsize=(20,8))
axes[0].plot(x,y, c='orange')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].plot(y,v, c='green')
axes[1].set_xlabel('y')
axes[1].set_ylabel('dy/dx')
```

```python
df = pd.DataFrame()
df['x'] = x
df['y'] = y
df['dy/dx'] = v
df
```
