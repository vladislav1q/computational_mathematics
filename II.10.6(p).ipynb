# 1) Сначала построим Matrix A и column f

```python
from math import sqrt
import numpy as np

a = [1,5,1]*48
b = [5,1] + [0]*48
c = [[0]*(x/3) + a[x:x+3] + [0]*(47 - x/3) for x in xrange(0, 48*3, 3)]
d = [1]*50
A = np.vstack([b, c,d])
#A
```

# 3)Метод Зейделя

```python
def seidel(A, b, eps):
    
    n = len(A)
    x = [.0 for i in range(n)]

    converge = False
    print "The residue is:"
    while not converge:
        x_new = np.copy(x)
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
#Критерий останова!!!
        converge = sqrt(sum((x_new[i] - x[i]) ** 2 for i in range(n))) <= eps
# Вычисление и печать невязки
        residue = b - np.dot(A, x_new)
        print residue
        
        x = x_new

    return x
```

# 4) Метод Гаусса

## Данная функция не работает для решения произвольных систем методом гаусса, т.к. здесь реализован выбор первого элемента в столбце, а не наибольшего по модулю

```python
def gauss(A, b):
    n = len(A)
    A = np.array(A)
    A = A.astype(float)
    
    
    A = np.column_stack((A, b))
    
    for j in range(0, n):
        for i in range(j+1,n):
            if A[i][j] != 0:
                A[i] = A[i] - A[i][j]/A[j][j] * A[j]
                       
    x = np.array([.0 for i in range(n)])
    
    for i in range(n-1, -1, -1):
        x[i] = A[i][n]/A[i][i] - sum(A[i][k]/A[i][i]*x[k] for k in range(i+1, n))
    return x
```

# 5) Теперь посчитаем результат для двух методов и сравним его с тем что должно было получиться

```python
x = seidel(A, f, 10**-3)
print '\nresult is'
print x
print '\nerf is:'
print x - np.linalg.solve(A,f)
```

```python
x = gauss(A, f)
print '\nresult is'
print x
print '\nerf is:'
print x - np.linalg.solve(A,f)
```

## Видно, что даже метод гаусса(прямой метод) не дает точного решения данной системы из-за неточного представления числа в памяти компьютера

# 6) Найдем наибольшие и наименьшие собственные числа матрицы, определим число обусловленности

```python
n, v = np.linalg.eig(A)
print 'All eigenvalues'
print n

print 'Max eigenvalue:'
print max(n)
print 'Min eigenvalue:'
print min(n)

print 'The condition number of a matrix(frobenius norm):'
print np.linalg.cond(A, p = 'fro')
print 'The condition number of a matrix(first norm):'
print np.linalg.cond(A, p = 1)
print 'The condition number of a matrix(second norm):'
print np.linalg.cond(A, p = 2)
```

## Видно, что в зависимости от определения нормы мы получаем различные числа обусловленности для матрицы
