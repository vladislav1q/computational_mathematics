Вычислить интеграл с точность $10^{-6}$:
$$\int_{0}^{1} \frac{\sin x}{1+x^2} dx$$

Укажите и сравните различные методы.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Решать задачу будем применяя методы прямоугольников, трапеций и симпсона. Для подбора шага будем пользоваться методом Рунге.

```python
def f(x): return np.sin(x)/(1+x**2)
a = 0 
b = 1
```

```python
def rectangle(f, h):
    return h * sum(f[:-1])
def trapezoid(f, h):
    return 0.5 * h * (f[0] + 2 * sum(f[1:-1]) + f[-1])
def simpson(f, h):
    return h/3. * (f[0] + 4 * sum(f[1:-1:2]) + 2 * sum(f[2:-2:2]) + f[-1])
```

Шаг ограничим снизу 1e-05

```python
def runge(f, a, b, method, eps=1e-6):
    n = 4.
    if method == simpson:
        p = 4
    else: p = 2
    Ih = method(f(np.linspace(a, b, n+1)), (b-a) / n)
    while True:
        Ih_2 = method(f(np.linspace(a, b, 2*n+1)), (b-a) / (2*n))
        Dh_2 = (Ih - Ih_2) / (2**p - 1) 
        print('h = %4f, I(h) = %.16f, err(h) = %.6e' % ((b-a) / n,Ih_2, abs(Dh_2)))
        if abs(Dh_2) < eps:
            break
        n *= 2
        Ih = Ih_2
        if n > 100000: print('Too little h'); break
    return Ih_2 - Dh_2, np.log2(n/4)+1
```

$\textbf{Метод прямоугольников}$

```python
rung = runge(f, a, b, method=rectangle, eps = 1e-6)
print 'Итоговое значение:  {}'.format(rung[0])
print 'Количество итераций:  {}'.format(rung[1])
```

$\textbf{Метод трапеций}$

```python
rung = runge(f, a, b, method=trapezoid, eps = 1e-6)
print 'Итоговое значение:  {}'.format(rung[0])
print 'Количество итераций:  {}'.format(rung[1])
```

$\textbf{Метод Симпсона}$

```python
rung = runge(f, a, b, method=simpson, eps = 1e-6)
print 'Итоговое значение:  {}'.format(rung[0])
print 'Количество итераций:  {}'.format(rung[1])
```

Оценим время исполнения для каждого метода:

$\textbf{Метод прямоугольников}$

```python
%timeit runge(f, a, b, method=rectangle, eps = 1e-6)
```

$\textbf{Метод трапеций}$

```python
%timeit runge(f, a, b, method=trapezoid, eps = 1e-6)
```

$\textbf{Метод Симпсона}$

```python
%timeit runge(f, a, b, method=simpson, eps = 1e-6)
```
